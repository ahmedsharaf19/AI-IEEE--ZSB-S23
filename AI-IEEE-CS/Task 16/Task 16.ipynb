```python
data['Do you have any volunteering experience?'].value_counts().plot(kind = 'bar')
```

```python
zagEnd = data[data['If yes, what is your volunteering experience?'].str.contains('zag eng|Zag Eng|ZagEng|Zag_Eng')].shape[0]
zigZag = data[data['If yes, what is your volunteering experience?'].str.contains('zigzag|ZigZag|زيجزاج')].shape[0]
IEEE = data[data['If yes, what is your volunteering experience?'].str.contains('IEEE|ieee|Ieee')].shape[0]
resala = data[data['If yes, what is your volunteering experience?'].str.contains('Resala|رسالة')].shape[0]
zufs = data[data['If yes, what is your volunteering experience?'].str.contains('ZUFS')].shape[0]
zesu = data[data['If yes, what is your volunteering experience?'].str.contains('ZESU')].shape[0]
yly = data[data['If yes, what is your volunteering experience?'].str.contains('YLY')].shape[0]
res = data[data['If yes, what is your volunteering experience?'].str.contains('RES')].shape[0]
others = data.shape[0] - IEEE - zagEnd - zigZag - resala - zufs - zesu - yly - res
sizes = [IEEE, zagEnd, zigZag, resala, zufs, zesu, yly, res, others]

labels = ['IEEE', 'ZagEnd', 'ZigZag', 'Resala', 'ZUFS', 'ZESU', 'YLY', 'RES', 'Others']
plt.figure(figsize=(10,6))
plt.bar(labels, sizes)
```

```python
plt.bar(['Head', 'Member'], [data[data['If yes, what is your volunteering experience?'].str.contains('head|Head|Leader')].shape[0], data[data['If yes, what is your volunteering experience?'].str.contains('member|عضو|Member|Trainee')].shape[0]])
```

```python
sizes = [data[data['Faculty - Department'].str.contains('Electric|كهربا|electric|elec')].shape[0],
        data[data['Faculty - Department'].str.contains('Mechatron|mechatron|ميكاترون')].shape[0],
        data[data['Faculty - Department'].str.contains('mechan|Mechan|ميكان')].shape[0],
        data[data['Faculty - Department'].str.contains('Prep|prep|اعدادي')].shape[0],
        ]

labels = ['Electric', 'Mechatronics', 'Mechanical', 'Prepatory']
plt.bar(labels, sizes)
```

```python
import matplotlib.pyplot as plt
sizes = [data[data['How did you hear about us?'].str.contains('Faculty colleagues')].shape[0],
        data[data['How did you hear about us?'].str.contains('Friends')].shape[0],
        data[data['How did you hear about us?'].str.contains('Instagram')].shape[0],
        data[data['How did you hear about us?'].str.contains('Facebook')].shape[0]]

labels = ['Colleagues', 'Friends', 'Instagram', 'facebook']

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%')
```

```python
embedded = data[data['What do you seek to learn by joining Z Fast Team?'].str.contains('embedded|Embedded')].shape[0]
electric_car = data[data['What do you seek to learn by joining Z Fast Team?'].str.contains('electric car|السيرات الكهربية|electrical vehicle')].shape[0]
mech = data[data['What do you seek to learn by joining Z Fast Team?'].str.contains('الميكاني')].shape[0]
others = data.shape[0] - embedded - electric_car - mech
sizes = [embedded, electric_car, mech, others]
labels = ['Embedded', 'Electric Car', 'Mechanics', 'Others']
plt.title('What do you seek to learn by joining Z Fast Team?')
plt.pie(sizes, labels = labels, autopct='%1.1f%%')
```
